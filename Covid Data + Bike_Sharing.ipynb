```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
who_time_series = pd.read_csv("WHO_time_series.csv")
who_time_series.head()
```

```python
who_time_series.dtypes
```

```python
who_time_series["Date_reported"] = pd.to_datetime(who_time_series["Date_reported"])
```

```python
who_time_series.dtypes
```

```python
italy = who_time_series[who_time_series["Country"] == "Italy"]
```

```python
italy.head(3)
```

```python
plt.plot(italy["Date_reported"], italy["Cumulative_cases"])
plt.title("Italy: Cumulative Reported Cases")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.show()

print("Generally, quantity that increase very quickly in the beginning and the it slows down more and more over time has a logarithmic growth")
```

```python
india = who_time_series[who_time_series["Country"] == "India"]

plt.plot(india["Date_reported"], india["Cumulative_cases"])
plt.title("India: Cumulative Reported Cases")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.show()

print("India Shows exponential growth for the data we have, but when the number of new cases will decrease, the growth will become logarithmic.")
```

## 

```python
def plotting(kuch_bhi):
    kuch_bhi=who_time_series[who_time_series["Country"]==kuch_bhi]
    plt.plot(kuch_bhi["Date_reported"],kuch_bhi["Cumulative_cases"])
    plt.title=(f"{kuch_bhi}","Cumulative Reported Cases")
    plt.xlabel("Date")
    plt.ylabel("Number of Cases")
    plt.show()
```

```python
plotting("India")
```

```python
def plot_cases(country_name):
    df = who_time_series[who_time_series["Country"] == country_name]

    plt.plot(df["Date_reported"], df["Cumulative_cases"])
    plt.title(f"{country_name}: Cumulative Reported Cases")
    plt.xlabel("Date")
    plt.ylabel("Number of Cases")
    plt.show()
```

```python
plot_cases("Poland")
print("Generally, a quantity that increase constantly over time has linear growth.")
```

```python
plot_cases("China")
```

```python
france = who_time_series[who_time_series["Country"] == "France"]
uk = who_time_series[who_time_series["Country"] == "The United Kingdom"]
pakistan = who_time_series[who_time_series["Country"] == "Pakistan"]
```

```python
plt.plot(france["Date_reported"], france["Cumulative_cases"], label = "France")
plt.plot(uk["Date_reported"], uk["Cumulative_cases"], label = "UK")
plt.plot(pakistan["Date_reported"], pakistan["Cumulative_cases"], label = "Pakistan")

plt.legend()
plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.show()
```

# Scatter Plot and Correlations

```python
bike_sharing = pd.read_csv("day.csv")
```
